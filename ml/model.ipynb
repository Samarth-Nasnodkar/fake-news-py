```python
import pandas as pd  
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split  
from sklearn.metrics import accuracy_score  
from sklearn.metrics import classification_report  
import re  
import string  
```

```python
df_fake = pd.read_csv("ml/data/Fake.csv")
df_true = pd.read_csv("ml/data/True.csv")
```

```python
df_fake.head()
```

```python
df_true.head()
```

```python
df_fake['target'] = 0
df_true['target'] = 1
```

```python
df_fake.shape, df_true.shape
```

```python
val_fake = df_fake.tail(50)
val_true = df_true.tail(50)

for i in range(df_fake.shape[0] - 1, df_fake.shape[0] - 51, -1):
  df_fake.drop([i], axis = 0, inplace = True)

for i in range(df_true.shape[0] - 1, df_true.shape[0] - 51, -1):
  df_true.drop([i], axis = 0, inplace = True)

df_fake.shape, df_true.shape
```

```python
val_true.head()
```

```python
val_fake.head()
```

```python
df_val = pd.concat([val_fake, val_true], axis = 0)
df_val.to_csv("validation.csv")
```

```python
df_merged = pd.concat([df_fake, df_true], axis = 0)
df_merged.head()
```

```python
df_merged.columns
```

```python
df_merged.drop(["title", "subject", "date"], axis = 1, inplace = True)
```

```python
df_merged.isnull().sum()
```

```python
df_merged = df_merged.sample(frac = 1)
df_merged.reset_index(inplace = True)
df_merged.head()
```

```python
def processText(text: str) -> str:
  text = text.lower()
  text = re.sub('\[.*?\]', '', text)
  text = re.sub("\\W"," ",text) 
  text = re.sub('https?://\S+|www\.\S+', '', text)
  text = re.sub('<.*?>+', '', text)
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text = re.sub('\n', '', text)
  text = re.sub('\w*\d\w*', '', text)    
  return text
```

```python
df_merged['text'] = df_merged['text'].apply(processText)
```

```python
x = df_merged['text']
y = df_merged['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GBC = GradientBoostingClassifier(random_state = 0)
GBC.fit(xv_train, y_train)
```

```python
pred_gbc = GBC.predict(xv_test)
```

```python
GBC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(random_state = 0)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
```

```python
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
def output_label(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Not A Fake News"
```

```python
def testing(news):
  testing_news = {"text": [news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test['text'] = new_def_test['text'].apply(processText)
  new_x_test = new_def_test['text']
  new_xv_test = vectorization.transform(new_x_test)
  pred_LR = LR.predict(new_xv_test)
  pred_DT = DT.predict(new_xv_test)
  pred_GBC = GBC.predict(new_xv_test)
  pred_RFC = RFC.predict(new_xv_test)
  preds = [pred_LR[0], pred_DT[0], pred_GBC[0], pred_RFC[0]]
  return pred_LR
  # return np.argmax(np.bincount(preds))
  return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_label(pred_LR[0]),                                                                                                       
                                                                                                            output_label(pred_DT[0]), 
                                                                                                              output_label(pred_GBC[0]), 
                                                                                                              output_label(pred_RFC[0])))
```

```python
df_val = df_val.sample(frac=1)
df_val.reset_index(inplace = True)
df_val.head()
```

```python
corr_count = 0

for i in range(df_val.shape[0]):
  actual = df_val['target'].iloc[i]
  pred = testing(df_val['text'].iloc[i])
  if actual == pred:
    corr_count += 1

print("Accuracy: ", corr_count / df_val.shape[0])
```
